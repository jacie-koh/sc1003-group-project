```python
import csv
```

```python
mega_raw = []

with open('records.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        mega_raw.append(row)
```

```python
headers = mega_raw[0]
print(headers)
```

```python
by_tut_group = {}
for i in range(1,len(mega_raw)):
    if mega_raw[i][0] in by_tut_group:
        by_tut_group[mega_raw[i][0]].append(mega_raw[i][1:])
    else:
        by_tut_group[mega_raw[i][0]] = []
        by_tut_group[mega_raw[i][0]].append(mega_raw[i][1:])


print(len(by_tut_group))
```

```python
GPA_explorings = []
for i in range(1,len(mega_raw)):
    GPA_explorings.append(mega_raw[i][-1])

print(GPA_explorings)
print(len(GPA_explorings))
#avg 
GPA_explorings = list(map(lambda x: float(x),GPA_explorings))
mean = sum(GPA_explorings)/len(GPA_explorings)
print(f"MEAN: {mean}")

#var
sum_squared = list(map(lambda x: float(x)**2,GPA_explorings))
print(sum_squared)
variance = sum(sum_squared)/len(GPA_explorings) - mean**2
print(f"VARIANCE: {variance}")
```

```python
def find_mean_var_per_group(group_gpa):
    #avg 
    group_gpa = list(map(lambda x: float(x),group_gpa))
    mean = sum(group_gpa)/len(group_gpa)
    print(f"MEAN: {mean}")

    #var
    sum_squared = list(map(lambda x: float(x)**2,group_gpa))
    # print(sum_squared)
    variance = sum(sum_squared)/len(group_gpa) - mean**2
    print(f"VARIANCE: {variance}")
```

```python
for i in by_tut_group:
    tut_group = by_tut_group[i]
    tut_GAP = []
    for person in tut_group:
        tut_GAP.append(person[-1])
    print(i)
    find_mean_var_per_group(tut_GAP)
```
