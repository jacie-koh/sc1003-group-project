```python
import csv
from collections import defaultdict
```

```python
with open('records.csv', mode='r') as file:
    reader = csv.DictReader(file)
    data = [row for row in reader]
```

```python
from collections import defaultdict

def divide_into_groups(students):
    # Divide by gender
    males = [student for student in students if student['Gender'] == 'Male']
    females = [student for student in students if student['Gender'] == 'Female']
    
    # Sort by CGPS
    males = sorted(males, key=lambda x: float(x['CGPA']), reverse=True)
    females = sorted(females, key=lambda x: float(x['CGPA']), reverse=True)
    
    groups = defaultdict(list)
    
    # Checks whether the school already exists in the group
    def can_add_to_group(group, student):
        schools_in_group = [s['School'] for s in group]
        return schools_in_group.count(student['School']) < 1

    # Parameters
    group_count = 1
    group = []
    group_size = 4
    
    # To alternate in picking the highest and lowest CPGA
    from_start = True

    while males or females:
        if len(group) < group_size:  # adds if group not full
            # Alternate between adding males and females
            if len([s for s in group if s['Gender'] == 'Male']) < group_size // 2 and males:
                if from_start:
                    male_candidate = males.pop(0)
                else:
                    male_candidate = males.pop()
    
                if can_add_to_group(group, male_candidate):
                    group.append(male_candidate)
                # In case school already exists, checks for the next possible value to add
                else:
                    if from_start:
                        for i in range(len(males)):
                            if can_add_to_group(group, males[i]):
                                temp = males.pop(i) 
                                #returns the student previously picked
                                males.insert(0, male_candidate)
                                male_candidate = temp
                                break
                    else:
                        for i in range(len(males)-1, -1, -1):                            
                            if can_add_to_group(group, males[i]):
                                temp = males.pop(i)        
                                males.append(male_candidate)
                                male_candidate = temp
                                break
                                
                    group.append(male_candidate)
    
            if len([s for s in group if s['Gender'] == 'Female']) < group_size // 2 and females:
                if from_start:
                    female_candidate = females.pop(0)
                else:
                    female_candidate = females.pop()
    
                if can_add_to_group(group, female_candidate):
                    group.append(female_candidate)
                else:
                    if from_start:
                        for i in range(len(females)):
                            if can_add_to_group(group, females[i]):
                                temp = females.pop(i)
                                females.insert(0, female_candidate)
                                female_candidate = temp
                                break
                    else:
                        for i in range(len(females)-1, -1, -1):
                            if can_add_to_group(group, females[i]):
                                temp = females.pop(i)
                                females.append(female_candidate)
                                female_candidate = temp
                                break
                                
                    group.append(female_candidate)
    
            # Switch between picking from the start and end
            from_start = not from_start
    
        else:
            # Saves group and resets for a new group
            groups[f'Group-{group_count}'] = group
            group = []
            group_count += 1
            
            # Break if all students are assigned
            if not males or not females:
                break
            # Break after 1200 groups are formed
            if group_count == 1201:
                break

    # Create a list for the remaining median students
    remaining = males + females

    # Assign the remaining students to the groups
    for group in groups:
        groups[group].append(remaining.pop())

    return groups
```

```python
# Divide students
groups = divide_into_groups(data)

# Print the results
for group_name, group_members in groups.items():
    print(f"{group_name}:")
    for member in group_members:
        print(f"  {member['Name']} - {member['School']} - {member['Gender']} - {member['CGPA']}")
```
